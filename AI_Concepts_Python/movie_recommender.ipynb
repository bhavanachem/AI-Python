```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer #converts words to tokens, frequency
from sklearn.metrics.pairwise import cosine_similarity

import os
work_dir = "/content/drive/MyDrive/SKILLIT Courses/AI Level 2/Clustering/Movie_recommendation"
```

```python
os.chdir(work_dir)
df = pd.read_csv("movie_dataset.csv")
print(df.columns)
```

Columns / Attributes of movie

```python
#helper functions

def get_title_from_index(index):
  return df[df.index == index]["title"].values[0]

def get_index_from_title(title):
  return df[df.title==title]["index"].values[0]

  #Select Features

  features = ["keywords", "cast", "genres", "director"]

  for feature in features:
    df[feature] = df[feature].fillna("")

#coombining features

def combine_features(row):
  try:
    text = row['keywords'] + ' ' + row["cast"] + " " +row["genres"] + " " + row["director"]
    return text;
  except:
    print("error: ", row["index"])
```

```python
df['combined_features'] = df.apply(combine_features, axis = 1) # AXIS = 1 results in rows
df['combined_features']
```

Create count matrix from combined features

```python
# making computer understand the words

cv = CountVectorizer()
df["combined_features"] = df["combined_features"].astype(str)

count_matrix = cv.fit_transform(df["combined_features"])

cosine_sim = cosine_similarity(count_matrix)

```

```python
def movie_finder():
  movie_user_likes = input("which is your favotire movie?")
  movie_index = get_index_from_title(movie_user_likes)
  similar_movies = list(enumerate(cosine_sim[movie_index]))
  sorted_similar_movies = sorted(similar_movies, key=lambda x:x[1], reverse = True)
  i=0
  print("you might also like these movies: ")
  for movie in sorted_similar_movies:
    print(get_title_from_index(movie[0]))
    i=i+1
    if i>5:
      break;

movie_finder()
```
