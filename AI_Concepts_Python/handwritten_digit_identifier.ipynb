```python
import sklearn
from sklearn.datasets import fetch_openml #for reading files in datasets
import random
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

Load data

```python
mnist = fetch_openml('mnist_784') # dataset with handwritten digits
mnist
```

```python
x,y =  mnist['data'], mnist['target']

#random_digit = x[random.randint(1,10000)] # random num between 1 and 10,0000
#some_random_digit = random_digit.reshape(28,28) #one image
#plt.imshow(some_random_digit, cmap = matplotlib.cm.binary)  #ploting image
```

```python
x_train, x_test = x[:60000].to_numpy(), x[60000:70000].to_numpy() # Convert to NumPy arrays
y_train, y_test = y[:60000].to_numpy(), y[60000:70000].to_numpy() # Convert to NumPy arrays

shuffle_index = np.random.permutation(60000)
x_train, y_train = x_train[shuffle_index], y_train[shuffle_index] # Now this will work correctly
```

Define Model

```python
model = LogisticRegression(tol=0.1, max_iter = 2000) #tol is tolerance for stopping criteria
    #max_iter = max number of iterations taken

model.fit(x_train, y_train)
```

```python
test_num = int(input("Select a number between 1 and 10,000 to select image: "))
test_image_data = x_test[test_num]
test_image = test_image_data.reshape(28,28)

prediction = model.predict([test_image_data])

plt.imshow(test_image, cmap = matplotlib.cm.binary)
plt.title(f'Selected Image Number: {test_num}')
plt.xlabel(f"The image has been classified as: {prediction[0]}") # Changed 'predication' to 'prediction'
```
